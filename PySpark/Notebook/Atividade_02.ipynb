1. Crie um banco de dados no DW do Spark chamado VendasVarejo , e persista todas as tabelas neste banco de dados.

```
clientes = spark.read.load("/FileStore/tables/Clientes.parquet", format="parquet", sep=",", inferSchema=False, header=False)

vendas = spark.read.load("/FileStore/tables/Vendas.parquet", format="parquet", sep=",", inferSchema=False, header=False)

itensvendas = spark.read.load("/FileStore/tables/ItensVendas.parquet", format="parquet", sep=",", inferSchema=False, header=False)

produtos = spark.read.load("/FileStore/tables/Produtos.parquet", format="parquet", sep=",", inferSchema=False, header=False)

vendedores = spark.read.load("/FileStore/tables/Vendedores.parquet", format="parquet", sep=",", inferSchema=False, header=False)
```

```
clientes.show(3)
vendas.show(3)
itensvendas.show(3)
produtos.show(3)
vendedores.show(3)
```

Para criar o DW precisa criar o BD

```
spark.sql("create database VendasVarejo")
```

```
spark.sql("use VendasVarejo")
```

Salvar os DF criados neste BD

```
clientes.write.saveAsTable("clientes")
vendas.write.saveAsTable("vendas")
itensvendas.write.saveAsTable("itensvendas")
produtos.write.saveAsTable("produtos")
vendedores.write.saveAsTable("vendedores")
```

2. Crie uma consulta que mostre de cada item vendido: Nome do Cliente, Data da Venda,Produto, Vendedor e Valor Total do item.

```
%sql
use vendasvarejo;
```

```
%sql

select c.cliente, v.Data, p.Produto, vd.Vendedor, iv.ValorTotal
from itensvendas iv
inner join produtos p on (p.Produtoid = iv.Produtoid)
inner join vendas v on (v.VendasID = iv.VendasID)
inner join vendedores vd on (vd.vendedorID = v.vendedorID)
inner join clientes c on (c.clienteid = v.clienteid);
```
