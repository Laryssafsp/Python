#Realizando Agrupamento das flores

Criando varável x com as colunas PetalLengthCm	PetalWidthCm
- pega os valores e transforma num array 

```python
x_petala = base_iris.iloc[:,[2,3]].values
x_petala[:10]
```

#Normalização os dados

Como os dados estão em uma escala muito diferente, precisamos normalizar os dados colocando no mesmo padrão e na mesma escala.
Como o Kmeans realiza os calculos baseados na distancia, devemos padronizar os dados afim de que um atreibuto não seja considerado mais importante do que o outro.

```python
  normalizar_dados = StandardScaler() 
  x_petala = normalizar_dados.fit_transform(x_petala) #atribuir o fit_transform na variável
  x_petala[:10] #limitado a 10 registros
```

#Calculando o número de Cluster

Para calcular o número de cluster vamos utilizar o método do cotovelo
WCSS é a soma da distancia entre cada ponto e o centroide em um cluster

```python
wcss_petala = [] # criar lista vazia
for i in range(1,11): # inicia-se com 01 pois não existe 0 clusters
  kmeans_petala = KMeans(n_clusters=i, random_state=0) # Executa o Kmeans para todos os clusters e random_state = o para fixar e obter os mesmo resultados
  kmeans_petala.fit(x_petala) # realiza o treinamento
  wcss_petala.append(kmeans_petala.inertia_)
```

Visualizando os valores de wcss

```python
for i in range(len(wcss_petala)):
  print("Cluster:", i, "- valor do wcss", wcss_petala[i])
```

# Criando gráfico para melhor visualização

```python
grafico_colovelo_petala = px.line( x=range(1,11), y=wcss_petala)
grafico_colovelo_petala.show()
```

```python
kmeans_petala = KMeans(n_clusters=3, random_state = 0)
label_cluster_petala = kmeans_petala.fit_predict(x_petala)
```

Verificando e classificação dos clusters

```python
label_cluster_petala
```

Chamando o centroides dos CLusters

```python
centroides_petala = kmeans_petala.cluster_centers_
centroides_petala 
```

# Grafico de agrupamento das caracterísicas do tamanho e comprimento das pétalas

```python
grafico_petala = px.scatter(x = x_petala[:,0], y = x_petala[:,1], color = label_cluster_petala)
grafico_centroide_petala = px.scatter(x = centroides_petala[:,0], y = centroides_petala[:,1], size = [7,7,7])
grafico_final_petala = go.Figure(data = grafico_petala.data + grafico_centroide_petala.data)
grafico_final_petala.show()
```
