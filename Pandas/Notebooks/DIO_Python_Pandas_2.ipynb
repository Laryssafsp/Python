## Leitura do Arquivo

```python
df.tail()
```

```python
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

## **Tratando valores faltantes**



---



Consultando linhas com valores faltantes


```python
df.isnull().sum() # se for nulo, traga o soma
```

Substituindo os valores nulos pela média


```python
df["Vendas"].mean()
```

```python
# na coluna 'venda', substituir os valores nulos com a média da col 'venda'
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True) # inplace=True para modificar em memória
```

Verificando se temos Valores Nulos

```python
df.isnull().sum()
```

Substituindo os valores nulos por zero


```python
df["Vendas"].fillna(0, inplace=True) # inplace=True para modificar em memória
```

## Apagando as linhas com valores nulos


```python
df.dropna(inplace=True)
```

Apagando as linhas com valores nulos com base apenas em 1 coluna


```python
df.dropna(subset=["Vendas"], inplace=True)
```

Removendo linhas que estejam com valores faltantes em todas as colunas


```python
df.dropna(how="all", inplace=True)
```

 Qual a Receita total?

Qual o custo Total?

```python
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) 
#Criando a coluna de custo
```

Qual o custo Total?

```python
round(df["custo"].sum(), 2)
# arredondamento
```

 Agora que temos a receita e custo e o total, podemos achar o Lucro total
 Vamos criar uma coluna de Lucro que será Receita - Custo

```python
df["lucro"]  = df["Valor Venda"] - df["custo"] 
```

```python
df.head(1)
```

Total Lucro

```python
round(df["lucro"].sum(),2)
```

##Criando uma coluna com total de dias para enviar o produto
